```python
import pandas as pd
```

```python
df = pd.read_excel('saleslistasoj.xlsx', index_col=False)
```

```python
df.head(10)
```

```python
df_new = df.rename(columns={'Bhole Baba Order Suppliers': 'Date','Unnamed: 1': 'Vch/Bill No', 'Unnamed: 2': 'Particulars', 'Unnamed: 3': 'Item Details',  'Unnamed: 4': 'Qty', 'Unnamed: 5': 'Unit'})
```

```python
df_new.head(10)
df_new = df_new[3:len(df_new.index)-2]
```

```python
df_new = df_new.reset_index()
del df_new['index']
```

```python
df_new.head(10)
```

```python
df_new.isnull()
```

```python
df_new = df_new.fillna(0)
```

```python
for a in range(0,len(df_new.index)):
    if df_new['Date'][a] == 0:
        df_new['Date'][a] = df_new['Date'][a-1]
```

```python
df_new['Vch/Bill No'] = df_new['Vch/Bill No'].str.replace(' ', 'x')
```

```python
for a in range(0,len(df_new.index)):
    if df_new['Vch/Bill No'][a] == 'x':
        df_new['Vch/Bill No'][a] = df_new['Vch/Bill No'][a-1]
```

```python
df_new['Particulars'] = df_new['Particulars'].str.replace(' ', 'x')
```

```python
for a in range(0,len(df_new.index)):
    if df_new['Particulars'][a] == 'x':
        df_new['Particulars'][a] = df_new['Particulars'][a-1]
```
